```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('bbc_News_Dataset.csv')
df.head()
print(df.shape)
```

```python
pd.unique(df['category'])
```

```python
from nltk.tokenize import word_tokenize

tokenized_titles = []

for title in df['title']:
    if pd.isna(title):
        tokenized_titles.append([])  
    else:
        tokenized_titles.append(word_tokenize(title))
```

```python

for title in tokenized_titles:
    print(title)
```

```python
import pickle
import os

file = 'pklFiles/tokenized_titles.pkl'

os.makedirs(os.path.dirname(file), exist_ok=True)

with open(file, 'wb') as fileobj:
    pickle.dump(tokenized_titles, fileobj)
```

```python
import pickle

file = 'pklFiles/tokenized_titles.pkl'
fileobj = open(file, 'wb')
pickle.dump(tokenized_titles, fileobj)
fileobj.close()
```

```python


from nltk.corpus import stopwords
import string

stop_words = set(stopwords.words('english'))
print("Stop Words: ", stop_words)


punctuations = set(string.punctuation)
print("Punctuations: ", punctuations)


filtered_titles = []

for title in tokenized_titles:
    filtered_title = []
    for word in title:
        if word not in stop_words and word not in punctuations:
            filtered_title.append(word)
    filtered_titles.append(filtered_title)



print("\nFiltered titles: \n", filtered_titles)

```

```python
file = 'pklFiles/filtered_titles.pkl'
fileobj = open(file, 'wb')
pickle.dump(filtered_titles, fileobj)
fileobj.close()
```

```python
from nltk.stem.porter import PorterStemmer

porter = PorterStemmer()

stemmed_headlines = []

for title in filtered_titles:
    stemmed_title = []
    for word in title:
        stemmed_title.append(porter.stem(word))
    stemmed_headlines.append(" ".join(stemmed_title))


print("\nStemmed Headlines: \n", stemmed_headlines)
```

```python
file = 'pklFiles/filtered_titles.pkl'
fileobj = open(file, 'wb')
pickle.dump(filtered_titles, fileobj)
fileobj.close()
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
df['ENCODED_CATEGORY'] = labelencoder.fit_transform(df['category'])

category_mapping = dict(zip(labelencoder.classes_, labelencoder.transform(labelencoder.classes_)))

print("Category Encoding Mapping:")
for category, encoded_value in category_mapping.items():
    print(f"{category}: {encoded_value}")

```
